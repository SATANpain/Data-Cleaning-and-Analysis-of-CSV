```python
import matplotlib.pyplot as plt
```

```python
years = [2015, 2016, 2017, 2018, 2019, 2020]
sales = [200, 300, 400, 550, 600, 700]

plt.plot(years, sales, marker='s',  color='r', label='Sales', linestyle='-')

plt.xlabel('year')
plt.ylabel('sales in (usd)')
plt.title("yearly  sales growth")
plt.legend()

plt.show
```

```python
import matplotlib.pyplot as plt

categories= ['electronics', 'clothing', 'home decor', 'books']
sales = [500,700,500,300]

plt.bar(categories, sales, color=['blue', 'green', 'red', 'yellow'])
plt.xlabel('product catogor')
plt.ylabel('sales in usd')
plt.title('sales by cateogry')

plt.show()          
                                  
                                  
```

```python
import seaborn as sns 
import pandas as pd 

data = {
    'categories': ['electronics', 'clothing', 'home decor', 'books'],
    'sales': [500,700,500,300],
}

df= pd.DataFrame(data)


sns.barplot(x='categories', y='sales', data=df, palette='coolwarm')
plt.title('sales by category seaborn')
plt.show()

    
```
